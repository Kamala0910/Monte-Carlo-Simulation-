#***Purpose:To simulate the value of pi using Monte Carlo Simulation Method:***
Our interest here is to estimate the value of pi.
Let (x1, x2) be a point, we will try to check the following condition,

if x^2 + y^2<=1:

increment the number of points that appears inside the circle.



#*Importing the required packages*

numpy-To perform a wide variety of mathematical operations on arrays.

matplotlib-To create different types of visualization like plots and charts.

random-to generate random numbers.

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from numpy.random import uniform
```

#*Setting up the required variables to store the values during the computations*

num_sample-to store the number of points to be generated.

Number of points(sample) is directly proportional to the accuracy of the estimation and time taken to compute.

linspace-a function of numpy to create an evenly spaced sequence in a specified interval.

A looping structure(for loop) is used to iterate over 1000 equally spaced values of i ranging from 0.0 to 1.0 generated using np.linspace.

Populating the values of x,y with 1000 points.
The currently generated value of x coordinate is stored in x and the corresponding y value is calculated using the unit circle equation:x^2+y^2=1.

Try giving values for num_sample as 10,100,1000,100000,10000000 to note some interesting patten=rns in plot and the impact on the value of pi.

```python
num_sample=1000
x=[]
y=[]

for i in np.linspace(0.0,1.0,100):
    x.append(i)
    y.append(np.sqrt(1.0-i*i))
```

#*Printing the estimated value of pi(here p)*
np.arange to generate a sequence of numbers from 0 to num_sample-1.

np.random.uniform to generate points x1,x2 within the unit square.

rx and ry are the lists that keeps track of the generated points.

The conditional statement (if) checks if the point lies inside the circle or not. If yes, it increments the variable val by 1 to keep track of the actual points that lie inside the circle.

It is done by using the condition, x^2+y^2<=r^2(here r=1).

We know estimate the value of pi which is the ratio of the number of points inside the unit circle (val) to the total number of points generated (num_sample) and multiply it to 4.


```python
rx=[]
ry=[]
val=0

for i in np.arange(num_sample):
    x1=np.random.uniform(0.0,1.0)
    x2=np.random.uniform(0.0,1.0)
    rx.append(x1)
    ry.append(x2)
    if(((x1*x1)+(x2*x2))<1.0):
        val+=1
p=float(val)/float(num_sample)*4.0
print("The estimated value of pi",p)
```

#*Generating a plot to visualize the results of the simulation for estimating the value of pi*

```python
fig,ax=plt.subplots()
ax.plot(x,y,linewidth=5)
ax.plot(rx,ry,'rx')
ax.set(xlabel='n value',ylabel='y value',title='Monte Carlo simulation to estimate value of Pi')
ax.set_xlim(0.0,1.0)
ax.set_ylim(0.0,1.0)
ax.set_aspect('equal')
ax.grid()
fig.savefig("inference.png")
plt.show()
```
